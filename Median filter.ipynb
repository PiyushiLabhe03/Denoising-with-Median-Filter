```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load an image (use the correct path to your image)
image_path = "C:/Users/Piyushi Labhe/Downloads/sharpened_image.tiff"
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Check if the image was successfully loaded
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Convert the image from BGR to RGB for display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply Median Filtering with a larger ksize to reduce more noise
    median_filtered = cv2.medianBlur(image, ksize=7)

    # Sharpening the median-filtered image using a stronger unsharp mask technique
    sharpen_kernel = np.array([[-1, -1, -1],
                               [-1, 9, -1],
                               [-1, -1, -1]])
    sharpened = cv2.filter2D(median_filtered, -1, sharpen_kernel)

    # Convert the sharpened image to RGB for display
    sharpened_rgb = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB)

    # Display the images using Matplotlib
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title('Original Image')
    plt.imshow(image_rgb)
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title('Median Filtered (ksize=7)')
    plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title('Sharpened Median Filtered Image')
    plt.imshow(sharpened_rgb)
    plt.axis('off')

    plt.show()
```
